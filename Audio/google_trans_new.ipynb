```
# Translation
from google_trans_new import google_translator 

translator = google_translator()
translate_text = translator.translate('สวัสดีจีน',lang_tgt='en')  
print(translate_text)
```

```
# Language Detection
detector = google_translator()  
detect_result = detector.detect('สวัสดีจีน')

print(detect_result)
```

```
# Pronounce
#translator  = google_translator()  
pronounce = translator.translate('สวัสดีจีน',lang_src='th',lang_tgt='zh',pronounce=True)  
print(pronounce)
```

```
# Text To Speech
from gtts import gTTS

from IPython.display import Audio

original_sentence = 'Buongiorno a tutti, come va oggi?'

translate_text = translator.translate(original_sentence,lang_tgt='en')  

tts = gTTS(translate_text)
tts.save('trans.mp3')
```

```
# Play the audio file
Audio('trans.mp3',autoplay=True)
```

```
# Changing the intonation (Italian)
tts = gTTS(translate_text,lang='it')
tts.save('trans.mp3')

Audio('trans.mp3',autoplay=True)
```

```
# Changing the intonation (Indian)
tts = gTTS(translate_text,lang='hi')
tts.save('trans.mp3')

Audio('trans.mp3',autoplay=True)
```

```
# Changing the intonation (French)
tts = gTTS(translate_text,lang='fr')
tts.save('trans.mp3')

Audio('trans.mp3',autoplay=True)
```
