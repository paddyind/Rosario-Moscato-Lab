# pyttsx3

[Link to PyPi](https://pypi.org/project/pyttsx3/)

**text-to-speech** conversion library in Python that works offline and is compatible with both Python 2 and 3.

### Speaking text

```python
import pyttsx3
#engine = pyttsx3.init("sapi5")
engine = pyttsx3.init()
engine.say('The quick brown fox jumped over the lazy dog.')
engine.runAndWait()
```

### Saving voice to a file

```python
import pyttsx3
engine = pyttsx3.init()
engine.save_to_file('Hello World' , 'hello_world.mp3')
engine.runAndWait()
```

### Changing Rate

```python
#import pyttsx3
#engine = pyttsx3.init() # object creation

rate = engine.getProperty('rate')   # getting details of current speaking rate -the base value is 200.
print("Rate:",rate)                 #printing current voice rate
```

```python
engine.setProperty('rate', 80)     # setting up new voice rate

engine.say('The quick brown fox jumped over the lazy dog.')
engine.runAndWait()
```

### Changing Volume

```python
volume = engine.getProperty('volume')   #getting to know current volume level (min=0 and max=1)
print ("Volume:",volume)                          #printing current volume level
```

```python
engine.setProperty('volume',0.2)    # setting up volume level  between 0 and 1

engine.say('The quick brown fox jumped over the lazy dog.')
engine.runAndWait()
```

### Changing Voice

```python
voices = engine.getProperty('voices')       #getting details of current voice
print(len(voices))
#print(voices)
```

```python
i=0
for voice in voices:
    print("Number:",i)
    print("Voice:",voice.name)
    print(" - ID:",voice.id)
    print(" - Languages:",voice.languages)
    print(" - Gender:",voice.gender)
    print(" - Age:",voice.age)
    print("\n")
    i+=1
```

```python
engine.setProperty("voice", voices[36].id)

engine.setProperty('rate', 200)     # setting up new voice rate
engine.setProperty('volume',1.0)    # setting up volume level  between 0 and 1

engine.say('Buongiorno a tutti questa è la lingua italiana')
engine.runAndWait()
```
